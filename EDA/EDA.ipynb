```python

df['9_ema'] = df['close'].ewm(span=9, adjust=False).mean()
df['21_ema'] = df['close'].ewm(span=21, adjust=False).mean()
df['200_ema'] = df['close'].ewm(span=200, adjust=False).mean()


plt.figure(figsize=(12, 16))

# Plotting the close price and its 9-day EMA
plt.subplot(4, 1, 1)
plt.plot(df['datetime'], df['close'], label='Close Price', color='blue')
plt.plot(df['datetime'], df['9_ema'], label='9-day EMA', color='red', linestyle='--')
plt.title('Cryptocurrency Close Price and 9-day EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Plotting the close price and its 21-day EMA
plt.subplot(4, 1, 2)
plt.plot(df['datetime'], df['close'], label='Close Price', color='blue')
plt.plot(df['datetime'], df['21_ema'], label='21-day EMA', color='green', linestyle='--')
plt.title('Cryptocurrency Close Price and 21-day EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Plotting the close price and its 50-day EMA
plt.subplot(4, 1, 3)
plt.plot(df['datetime'], df['close'], label='Close Price', color='blue')
plt.plot(df['datetime'], df['50_ema'], label='50-day EMA', color='orange', linestyle='--')
plt.title('Cryptocurrency Close Price and 50-day EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Plotting the close price and its 200-day EMA
plt.subplot(4, 1, 4)
plt.plot(df['datetime'], df['close'], label='Close Price', color='blue')
plt.plot(df['datetime'], df['200_ema'], label='200-day EMA', color='purple', linestyle='--')
plt.title('Cryptocurrency Close Price and 200-day EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Examine market dynamics using metrics like nrpl and nupl
plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['nupl'], label='Nupl')
plt.title('Market Dynamics: Nupl')
plt.xlabel('Datetime')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python

df['nupl_nrpl_diff'] = df['nupl'] - df['nrpl']


bullish_condition = (df['nupl_nrpl_diff'] > 0)
bearish_condition = (df['nupl_nrpl_diff'] < 0)

# Plotting the difference over time
plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['nupl_nrpl_diff'], label='NUPL - NRPL')

# Mark bullish sentiment periods
plt.scatter(df[bullish_condition]['datetime'], df[bullish_condition]['nupl_nrpl_diff'], color='green', label='Bullish Sentiment')

# Mark bearish sentiment periods
plt.scatter(df[bearish_condition]['datetime'], df[bearish_condition]['nupl_nrpl_diff'], color='red', label='Bearish Sentiment')

plt.title('Difference between NUPL and NRPL Over Time')
plt.xlabel('Datetime')
plt.ylabel('Difference')
plt.legend()
plt.grid(True)
plt.show()
```
